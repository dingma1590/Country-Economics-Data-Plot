# Project Presentation

##### Give an overview of your project. What problem are you solving? Where is the information from? Why is it interesting to you?
- The aim of our project is to plot the relationship between unemployment rate and GDP on a scatterplot, and visualize the distribution of all the countries in terms of those two data.
- The source is a default one, from a GitHub page. 
- It's interesting because I've never looked at anything related to demographics, so I thought it might be interesting to see what I can find.

##### Explain how you used what you learned about systematic program design to solve the problem. What design choices did you have to make? Why did you make the choices that you made?
- One decision made was choosing the fields for data definitions. When we visualized the graph in planning phase, we realized that all we need from the csv file are the two data we're plotting (GDP & population), so those were the only two fields we had to make the program simpler.

##### Describe how you solved the problem. Provide visuals if possible, and definitely provide the graph or chart(s) that your program created.
- First, we used the HtDAP template to create the baseline (data definitions, read function)
- Then, we approached the program like a funnel; started with the most general functions and ended with the most specific
- Started with the function for showing a scatter plot > functions to produce lists for x and y axis > functions for finding the average > a helper function.
- In the end, we placed all those inner functions into one composite function (analyze function), which also finishes the main function.


##### Discuss which part of the project was the most difficult for you and why
- The most difficult part was probably finding the code to plot vertical/horizontal lines. Initially I tried somehow doing it like how line graphs are done. It either didn't work or it wasn't a true reference line that'd work for any data. So I looked in the guide and eventually found the way to plot it.
